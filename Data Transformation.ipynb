```python
#Loding and Rename
import pandas as pd
import seaborn as sns
import numpy as np
data=pd.read_csv('https://raw.githubusercontent.com/fazlyrabbi77/DataProcessing/master/real-estate.csv')
```

```python
data
```

```python
#Renameing columns with inplace
data.rename(columns={'NUM_BEDROOMS':'Bedrooms','NUM_BATH':'Bath'},inplace=True)
data.head()
```

```python
data.isnull().values.any()
```

```python
data.isnull().values.sum()
```

```python
#Showing null value in row
null_columns=data.columns[data.isnull().any()]
data[null_columns].isnull().sum()
```

```python
#Filling null values with specific Value
data['PID'].fillna(100005000,inplace=True)
data
```

```python
#Row wise data filtering
data.loc[2,"ST_NUM"]=197
data.loc[6,"ST_NUM"]=208
data
```

```python
#Unwanted value treatment
cnt=0
for row in data['OWN_OCCUPIED']:
    try:
        int(row)
        data.loc[cnt,'OWN_OCCUPIED']=np.nan
    except ValueError:
            pass
    cnt+=1
        
data        
```

```python
data['Bedrooms']=pd.to_numeric(data['Bedrooms'],errors='coerce')
data['Bath']=pd.to_numeric(data['Bath'],errors='coerce')
data['SQ_FT']=pd.to_numeric(data['SQ_FT'],errors='coerce')
data
```

```python
#Mod ar maddome nan value 
data['OWN_OCCUPIED'].fillna(data['OWN_OCCUPIED'].mode()[0],inplace=True)
data
```

```python
#Median 
#groupe by parameter check
data.groupby('Bedrooms')['SQ_FT'].median()
```

```python
#filling null with group by vparameter
data['SQ_FT']=data['SQ_FT'].fillna(data.groupby('Bedrooms')['SQ_FT'].transform('median'))
data['SQ_FT']=data['SQ_FT'].fillna(data['SQ_FT'].median())
data
```

```python
data.loc[2,'Bedrooms']=1
data.loc[5,'Bedrooms']=1
data.loc[8,'Bedrooms']=3
data
```

```python
#bfill & ffill
data['Bath']=data['Bath'].fillna(method='bfill')
data
```

```python
#Type converstion
#Converting streed number to int
data.PID=data.PID.astype('int64')
data.ST_NUM=data.ST_NUM.astype('int64')
data.Bedrooms=data.Bedrooms.astype('int64')
data.SQ_FT=data.SQ_FT.astype('int64')
data
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['ST_NAME']=le.fit_transform(data['ST_NAME'])
data
```

```python
#Use of mapping function
mapping={'Y':1,'N':2}
data['OWN_OCCUPIED']=data['OWN_OCCUPIED'].map(mapping)
data
```

```python
#One Hot Encoding for nominal data
data1=pd.get_dummies(data,columns=['OWN_OCCUPIED'])

data1
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
scaler.fit(data)
scaled= pd.DataFrame(scaler.transform(data),columns=data.columns)
scaled
```

```python
#Standert Scaler
import pandas as pd
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
data_scaledstd=pd.DataFrame(ss.fit_transform(data),columns=data.columns)
data_scaledstd
```

```python
#RobustScaler
from sklearn.preprocessing import RobustScaler
robust=RobustScaler()
robust_scaled_data=robust.fit_transform(data)
robust_scaled_data=pd.DataFrame(robust_scaled_data,columns=data.columns)
robust_scaled_data
```
